# Task 1 - Schema Analysis

Analyze the schema of all CSV files in the "/content/northdata_extracted/northdata/" directory to determine if they are identical. If any discrepancies exist, identify the files and the specific differences, and then summarize the findings.

## List all CSV files

### Subtask:
Identify and list all CSV files present in the '/content/northdata_extracted/northdata/' directory.


**Reasoning**:
To identify and list all CSV files, I will use the `os` module to navigate the specified directory and filter files based on their extension as per the instructions.



```python
import os

# Define the path to the directory containing the CSV files
directory_path = '/content/northdata_extracted/northdata/'

# Get a list of all files in the specified directory
all_files = os.listdir(directory_path)

# Filter the list to include only files that end with '.csv'
csv_files = [f for f in all_files if f.endswith('.csv')]

print(f"Found {len(csv_files)} CSV files in '{directory_path}':")
for csv_file in csv_files:
    print(csv_file)
```

**Reasoning**:
The subtask of listing all CSV files is complete. The next logical step is to read the schema (header) of each identified CSV file to prepare for schema comparison, as stated in the main task. This will allow us to determine if all CSV files have an identical schema.



```python
import pandas as pd
import os

# Dictionary to store schemas of all CSV files
csv_schemas = {}

# Loop through each CSV file and read its header
for csv_file_name in csv_files:
    file_path = os.path.join(directory_path, csv_file_name)
    try:
        # Read only the header (first row) to get column names, specifying latin1 encoding
        df_temp = pd.read_csv(file_path, nrows=0, encoding='latin1')
        csv_schemas[csv_file_name] = df_temp.columns.tolist()
    except Exception as e:
        print(f"Error reading schema for {csv_file_name}: {e}")

print(f"Successfully read schemas for {len(csv_schemas)} CSV files.")
# Display schema for the first few files to verify
for i, (file_name, schema) in enumerate(csv_schemas.items()):
    if i < 5:
        print(f"File: {file_name}, Schema: {schema}")
    else:
        break
```

### Compare schemas

**Reasoning**:
To compare the schemas, I will first identify the unique schemas present across all files. If there's only one unique schema, it means all files are identical. If there are multiple unique schemas, I will list each unique schema and the files associated with it to pinpoint discrepancies.

**Reasoning**:
Now that the schemas for all CSV files are loaded, I will compare them by identifying unique schemas and grouping files that share the same schema. This will help in summarizing the schema discrepancies as per the main task.



```python
unique_schemas = {}
for file_name, schema in csv_schemas.items():
    schema_tuple = tuple(schema)  # Convert list to tuple to make it hashable for dictionary keys
    if schema_tuple not in unique_schemas:
        unique_schemas[schema_tuple] = []
    unique_schemas[schema_tuple].append(file_name)

if len(unique_schemas) == 1:
    print("All CSV files have an identical schema.")
    # print(f"Identical Schema: {list(unique_schemas.keys())[0]}")
else:
    print(f"Found {len(unique_schemas)} distinct schemas among the CSV files.\n")
    for i, (schema, files) in enumerate(unique_schemas.items()):
        print(f"Schema {i + 1}: {list(schema)}")
        print(f"Files with this schema ({len(files)}): {files[:5]}... (and {len(files) - 5} more)\n")
```

**Reasoning**:
The previous step identified 3 distinct schemas. To fulfill the subtask of summarizing the findings and identifying specific differences, I need to break down these schemas into individual column names and then systematically compare them. This will allow for a clear articulation of where the schemas diverge.



```python
schema_keys = list(unique_schemas.keys())

# Extract and split the schemas into lists of individual column names
schema_1_cols = list(schema_keys[0])[0].split(';')
schema_2_cols = list(schema_keys[1])[0].split(';')
schema_3_cols = list(schema_keys[2])[0].split(';')

print("Schema 1 (first 5 columns):", schema_1_cols[:5])
print("Schema 2 (first 5 columns):", schema_2_cols[:5])
print("Schema 3 (first 5 columns):", schema_3_cols[:5])

# Find differences between Schema 1 and Schema 2
diff_1_2 = set(schema_1_cols) ^ set(schema_2_cols)
print(f"\nDifferences between Schema 1 and Schema 2: {diff_1_2}")

# Find differences between Schema 1 and Schema 3
diff_1_3 = set(schema_1_cols) ^ set(schema_3_cols)
print(f"Differences between Schema 1 and Schema 3: {diff_1_3}")

# Find differences between Schema 2 and Schema 3
diff_2_3 = set(schema_2_cols) ^ set(schema_3_cols)
print(f"Differences between Schema 2 and Schema 3: {diff_2_3}")
```

## Combine files for Schema 1

### Subtask:
Read all CSV files belonging to Schema 1, concatenate them into a single DataFrame, and save it as 'combined_schema_1.csv'. Make sure to handle the specific column names and encoding ('latin1') used for this schema.


**Reasoning**:
To combine all CSV files belonging to Schema 1, I will iterate through the file paths, read each file into a DataFrame using the specified separator and encoding, and then concatenate them into a single DataFrame as instructed.



```python
import pandas as pd
import os

# Assuming `unique_schemas`, `directory_path` are available from previous steps

# Get the list of files for Schema 1 (the first schema in unique_schemas)
# The schema_keys[0] holds the tuple of column names as a single string.
# The files are stored in unique_schemas[schema_keys[0]].
schema_keys = list(unique_schemas.keys())
schema_1_files = unique_schemas[schema_keys[0]]

# Initialize an empty list to store DataFrames for Schema 1
schema1_dfs = []

print(f"Combining {len(schema_1_files)} files for Schema 1...")

# Iterate through the file names associated with Schema 1
for file_name in schema_1_files:
    file_path = os.path.join(directory_path, file_name)
    try:
        # Read the CSV file, specifying semicolon as separator and latin1 encoding
        # pandas will infer the header row by default if it's the first row.
        df = pd.read_csv(file_path, sep=';', encoding='latin1')
        schema1_dfs.append(df)
    except Exception as e:
        print(f"Error reading {file_name}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
if schema1_dfs:
    combined_schema1_df = pd.concat(schema1_dfs, ignore_index=True)

    # Save the combined DataFrame to a new CSV file
    output_file_name = 'combined_schema_1.csv'
    combined_schema1_df.to_csv(output_file_name, index=False, encoding='latin1')

    print(f"Successfully created '{output_file_name}' with {len(combined_schema1_df)} rows.")
else:
    print("No files were successfully processed for Schema 1.")
```

**Reasoning**:
Now that the files for Schema 1 are combined, I will proceed to combine all CSV files belonging to Schema 2, following similar steps: reading each file into a DataFrame using the specified separator and encoding, and then concatenating them into a single DataFrame as instructed.



```python
import pandas as pd
import os

# Get the list of files for Schema 2 (the second schema in unique_schemas)
schema_keys = list(unique_schemas.keys())
schema_2_files = unique_schemas[schema_keys[1]]

# Initialize an empty list to store DataFrames for Schema 2
schema2_dfs = []

print(f"Combining {len(schema_2_files)} files for Schema 2...")

# Iterate through the file names associated with Schema 2
for file_name in schema_2_files:
    file_path = os.path.join(directory_path, file_name)
    try:
        # Read the CSV file, specifying semicolon as separator and latin1 encoding
        df = pd.read_csv(file_path, sep=';', encoding='latin1')
        schema2_dfs.append(df)
    except Exception as e:
        print(f"Error reading {file_name}: {e}")

# Concatenate all DataFrames in the list into a single DataFrame
if schema2_dfs:
    combined_schema2_df = pd.concat(schema2_dfs, ignore_index=True)

    # Save the combined DataFrame to a new CSV file
    output_file_name = 'combined_schema_2.csv'
    combined_schema2_df.to_csv(output_file_name, index=False, encoding='latin1')

    print(f"Successfully created '{output_file_name}' with {len(combined_schema2_df)} rows.")
else:
    print("No files were successfully processed for Schema 2.")
```
