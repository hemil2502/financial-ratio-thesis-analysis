**Cell 5 - Define ratio groups**

```python
# All 13 winsorized ratios
ALL_RATIOS = [
    "profit_margin_win",
    "roa_win",
    "roe_win",
    "equity_ratio_calc_win",
    "debt_ratio_win",
    "cash_to_assets_win",
    "receivables_to_assets_win",
    "revenue_per_employee_calc_win",
    "profit_per_employee_win",
    "asset_turnover_win",
    "debt_to_equity_ratio_win",
    "cash_ratio_win",
    "receivables_turnover_win"
]

# Main thesis ratios
CORE_RATIOS = [
    "profit_margin_win",
    "roa_win",
    "equity_ratio_calc_win",
    "debt_ratio_win",
    "cash_to_assets_win",
    "receivables_to_assets_win",
    "asset_turnover_win"
]

# Secondary ratios
SECONDARY_RATIOS = [
    "roe_win",
    "debt_to_equity_ratio_win",
    "cash_ratio_win",
    "receivables_turnover_win"
]

# Use only in sub-sample analysis because employee data is limited
EMPLOYEE_RATIOS = [
    "revenue_per_employee_calc_win",
    "profit_per_employee_win"
]

missing_ratios = [col for col in ALL_RATIOS if col not in df.columns]

if missing_ratios:
    print("Missing ratio columns:")
    print(missing_ratios)
else:
    print("All 13 ratio columns are available.")
```

**Cell 6 - Detect important columns**

```python
def pick_col(possible_names):
    """
    Finds the first available column from a list of possible column names.
    """
    for col in possible_names:
        if col in df.columns:
            return col
    return None


YEAR_COL = pick_col(["financial_year", "year", "Year"])
STATUS_COL = pick_col(["status_group", "status", "company_status"])
SECTOR_COL = pick_col(["wz_section", "WZ_section", "section", "wz_section_label"])
WZ4_COL = pick_col(["wz_code", "WZ_code", "wz_4digit", "wz4"])
FIN_SECTOR_COL = pick_col(["is_financial_sector", "financial_sector"])
ASSETS_COL = pick_col(["Bilanzsumme EUR", "bilanzsumme_eur", "total_assets"])
REVENUE_COL = pick_col(["Umsatz EUR", "umsatz_eur", "revenue"])

important_cols = {
    "YEAR_COL": YEAR_COL,
    "STATUS_COL": STATUS_COL,
    "SECTOR_COL": SECTOR_COL,
    "WZ4_COL": WZ4_COL,
    "FIN_SECTOR_COL": FIN_SECTOR_COL,
    "ASSETS_COL": ASSETS_COL,
    "REVENUE_COL": REVENUE_COL
}

important_cols
```

**Cell 7 - Create log variables if missing**

```python
# Create log assets if not already available
if "log_assets" not in df.columns and ASSETS_COL is not None:
    df["log_assets"] = np.where(df[ASSETS_COL] > 0, np.log(df[ASSETS_COL]), np.nan)

# Create log revenue if not already available
if "log_revenue" not in df.columns and REVENUE_COL is not None:
    df["log_revenue"] = np.where(df[REVENUE_COL] > 0, np.log(df[REVENUE_COL]), np.nan)

print("Log variables checked/created.")

available_log_cols = [col for col in ["log_assets", "log_revenue"] if col in df.columns]
display(df[available_log_cols].describe().T)
```

**Cell 9 - Ratio availiblity table**

```python
ratio_availability = []

for col in ALL_RATIOS:
    valid_n = df[col].notna().sum()
    missing_n = df[col].isna().sum()
    valid_pct = valid_n / len(df) * 100

    if col in CORE_RATIOS:
        group = "Core ratio"
    elif col in SECONDARY_RATIOS:
        group = "Secondary ratio"
    else:
        group = "Employee sub-sample ratio"

    ratio_availability.append({
        "ratio": col,
        "ratio_group": group,
        "valid_n": valid_n,
        "missing_n": missing_n,
        "valid_pct": valid_pct
    })

ratio_availability = pd.DataFrame(ratio_availability)

ratio_availability_path = os.path.join(OUT_DIR, "04_02_ratio_availability.csv")
ratio_availability.to_csv(ratio_availability_path, index=False)

display(ratio_availability)

print("Saved:", ratio_availability_path)
```

**Cell 10 - Thesis-ready descriptive statistics for all 13 ratios**

```python
def descriptive_table(data, columns):
    rows = []

    for col in columns:
        s = data[col].dropna()

        rows.append({
            "variable": col,
            "n": s.shape[0],
            "mean": s.mean(),
            "std": s.std(),
            "min": s.min(),
            "p5": s.quantile(0.05),
            "p25": s.quantile(0.25),
            "median": s.median(),
            "p75": s.quantile(0.75),
            "p95": s.quantile(0.95),
            "max": s.max()
        })

    return pd.DataFrame(rows)


desc_all_ratios = descriptive_table(df, ALL_RATIOS)

desc_path = os.path.join(OUT_DIR, "04_03_all_13_ratio_descriptive_statistics.csv")
desc_all_ratios.to_csv(desc_path, index=False)

display(desc_all_ratios)

print("Saved:", desc_path)
```

**Cell 11 - Compact core ratio summary for thesis writing**

```python
core_summary = desc_all_ratios[
    desc_all_ratios["variable"].isin(CORE_RATIOS)
].copy()

core_summary = core_summary[
    ["variable", "n", "mean", "p25", "median", "p75", "p95"]
]

core_summary_path = os.path.join(OUT_DIR, "04_04_core_ratio_summary_for_thesis.csv")
core_summary.to_csv(core_summary_path, index=False)

display(core_summary)

print("Saved:", core_summary_path)
```

**Cell 13 - Sector-level core ratio medians**

```python
if YEAR_COL is not None:
    year_core_medians = (
        df.groupby(YEAR_COL)[CORE_RATIOS]
        .median()
        .reset_index()
    )

    year_counts = (
        df.groupby(YEAR_COL)
        .size()
        .reset_index(name="n")
    )

    year_core_medians = year_counts.merge(year_core_medians, on=YEAR_COL, how="left")

    year_path = os.path.join(OUT_DIR, "04_05_year_level_core_ratio_medians.csv")
    year_core_medians.to_csv(year_path, index=False)

    display(year_core_medians)
    print("Saved:", year_path)
else:
    print("Year column not found.")
```

**Cell 14 - Status comparison summary**

```python
if STATUS_COL is not None:
    status_counts = (
        df.groupby(STATUS_COL)
        .size()
        .reset_index(name="n")
        .sort_values("n", ascending=False)
    )

    status_counts["share_pct"] = status_counts["n"] / len(df) * 100

    status_core_medians = (
        df.groupby(STATUS_COL)[CORE_RATIOS]
        .median()
        .reset_index()
    )

    status_summary = status_counts.merge(status_core_medians, on=STATUS_COL, how="left")

    status_path = os.path.join(OUT_DIR, "04_07_status_core_ratio_summary.csv")
    status_summary.to_csv(status_path, index=False)

    display(status_summary)
    print("Saved:", status_path)
else:
    print("Status column not found.")
```

**Cell 15 - Financial-sector sensitivity summary**

```python
if FIN_SECTOR_COL is not None:
    fin_counts = (
        df.groupby(FIN_SECTOR_COL)
        .size()
        .reset_index(name="n")
    )

    fin_counts["share_pct"] = fin_counts["n"] / len(df) * 100

    fin_core_medians = (
        df.groupby(FIN_SECTOR_COL)[CORE_RATIOS]
        .median()
        .reset_index()
    )

    fin_summary = fin_counts.merge(fin_core_medians, on=FIN_SECTOR_COL, how="left")

    fin_path = os.path.join(OUT_DIR, "04_08_financial_sector_sensitivity_summary.csv")
    fin_summary.to_csv(fin_path, index=False)

    display(fin_summary)
    print("Saved:", fin_path)
else:
    print("Financial-sector column not found.")
```

**Cell 16 - Spearman correlation warning table**

```python
corr_matrix = df[ALL_RATIOS].corr(method="spearman", min_periods=100)

high_corr_pairs = []

for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):
        var1 = corr_matrix.columns[i]
        var2 = corr_matrix.columns[j]
        corr_value = corr_matrix.iloc[i, j]

        if abs(corr_value) >= 0.70:
            high_corr_pairs.append({
                "variable_1": var1,
                "variable_2": var2,
                "spearman_corr": corr_value,
                "abs_corr": abs(corr_value)
            })

high_corr_pairs = pd.DataFrame(high_corr_pairs).sort_values("abs_corr", ascending=False)

corr_path = os.path.join(OUT_DIR, "04_09_high_correlation_warning_pairs.csv")
high_corr_pairs.to_csv(corr_path, index=False)

display(high_corr_pairs)

print("Saved:", corr_path)
```

**Cell 20 - Model sample availibility check**

```python
model_variables = [
    "roa_win",
    "profit_margin_win",
    "debt_ratio_win",
    "equity_ratio_calc_win",
    "cash_to_assets_win",
    "receivables_to_assets_win",
    "asset_turnover_win",
    "log_assets"
]

# Add controls if available
if YEAR_COL is not None:
    model_variables.append(YEAR_COL)

if SECTOR_COL is not None:
    model_variables.append(SECTOR_COL)

if FIN_SECTOR_COL is not None:
    model_variables.append(FIN_SECTOR_COL)

model_variables = list(dict.fromkeys(model_variables))

model_sample_check = []

for outcome in ["roa_win", "profit_margin_win"]:
    vars_needed = [
        outcome,
        "debt_ratio_win",
        "cash_to_assets_win",
        "receivables_to_assets_win",
        "asset_turnover_win",
        "log_assets"
    ]

    if YEAR_COL is not None:
        vars_needed.append(YEAR_COL)

    if SECTOR_COL is not None:
        vars_needed.append(SECTOR_COL)

    vars_needed = [v for v in vars_needed if v in df.columns]

    complete_n = df[vars_needed].dropna().shape[0]

    model_sample_check.append({
        "outcome": outcome,
        "variables_used": " + ".join(vars_needed),
        "complete_case_n": complete_n,
        "share_of_full_sample_pct": complete_n / len(df) * 100
    })

model_sample_check = pd.DataFrame(model_sample_check)

sample_check_path = os.path.join(OUT_DIR, "04_13_model_sample_availability_check.csv")
model_sample_check.to_csv(sample_check_path, index=False)

display(model_sample_check)

print("Saved:", sample_check_path)
```

**Cell 23 — Helper function for saving thesis figures**

```python
def save_fig(filename):
    """
    Saves the current matplotlib figure in high quality.
    """
    path_png = os.path.join(FIG_DIR, filename + ".png")
    path_pdf = os.path.join(FIG_DIR, filename + ".pdf")

    plt.tight_layout()
    plt.savefig(path_png, dpi=300, bbox_inches="tight")
    plt.savefig(path_pdf, bbox_inches="tight")

    print("Saved:")
    print(path_png)
    print(path_pdf)
```

```python
if SECTOR_COL is not None:
    sector_pm = (
        df.groupby(SECTOR_COL)
        .agg(
            n=("profit_margin_win", "count"),
            median_profit_margin=("profit_margin_win", "median")
        )
        .reset_index()
    )

    # Keep only reliable sectors
    sector_pm = sector_pm[sector_pm["n"] >= 500]
    sector_pm = sector_pm.sort_values("median_profit_margin")

    sector_pm_path = os.path.join(OUT_DIR, "04_17_sector_median_profit_margin_for_figure.csv")
    sector_pm.to_csv(sector_pm_path, index=False)

    plt.figure(figsize=(10, 6))
    plt.barh(sector_pm[SECTOR_COL].astype(str), sector_pm["median_profit_margin"])
    plt.xlabel("Median profit margin")
    plt.ylabel("WZ sector")
    plt.title("Median profit margin by industry sector")

    save_fig("fig_03_median_profit_margin_by_sector")
    plt.show()

    display(sector_pm)
    print("Saved table:", sector_pm_path)
else:
    print("Sector column not found.")
```

**Cell 27 - Visual 4: Industry financial profile heatmap**

```python
if SECTOR_COL is not None:
    sector_counts = df.groupby(SECTOR_COL).size()
    reliable_sectors = sector_counts[sector_counts >= 500].index

    sector_profile = (
        df[df[SECTOR_COL].isin(reliable_sectors)]
        .groupby(SECTOR_COL)[CORE_RATIOS]
        .median()
    )

    # Standardize columns so different ratio scales are comparable
    sector_profile_z = (
        sector_profile - sector_profile.mean()
    ) / sector_profile.std(ddof=0)

    profile_path = os.path.join(OUT_DIR, "04_18_sector_core_ratio_profile_heatmap_data.csv")
    sector_profile.to_csv(profile_path)

    plt.figure(figsize=(12, 7))
    plt.imshow(sector_profile_z, aspect="auto")
    plt.colorbar(label="Standardized sector median")

    plt.xticks(
        ticks=np.arange(len(sector_profile_z.columns)),
        labels=sector_profile_z.columns,
        rotation=45,
        ha="right"
    )

    plt.yticks(
        ticks=np.arange(len(sector_profile_z.index)),
        labels=sector_profile_z.index.astype(str)
    )

    plt.title("Industry financial profile based on core ratios")

    save_fig("fig_04_industry_financial_profile_heatmap")
    plt.show()

    display(sector_profile)
    print("Saved table:", profile_path)
else:
    print("Sector column not found.")
```
